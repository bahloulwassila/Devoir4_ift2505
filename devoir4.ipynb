## Exercice 2

a) Le centre analytique d'un ensemble est le point qui minimise la fonction barrière.

La fonction barrière de l'ensemble $Y = \{ y \,|\, A^T y \leq c \}$ est la somme négative des logarithmes de variables d'écart $s_j$. 

$-\sum_{j=1}^{n} \ln(s_j)$ où $s_j$ = $c_j$ - ($A^Ty)_j$ tq $s_j$ > 0 $\quad \text{pour tout } j$


Afin de minimiser la fonction barrière, nous devons mettre le problème de cet ensemble sous forme standard en introduisant les variabes d'éacart $s_j$ : 

$\implies$ $A^Ty + s = c$ ,   s > 0

Maximiser le produit des variables d'écart (noté $\prod_{}{}(s)$ ) équivaut à maximisier $\sum_{j=1}^{n} \ln(s_j)$. 


## Propriété de la fonction logarithme : 

Elle tend vers l'infini négatif lorsque son argument tend vers 0. 




Ainsi, minimiser $-\sum_{j=1}^{n} \ln(s_j)$ entraine les variables d'écart à être éloignées de 0, tout en optimisant la solution. 

Donc maximiser $\prod_{}{}(s)$ $\implies$ maximiser $ln(\prod_{}{}(s))$

Ainsi le problème d'optimisation devient : max $ln(\prod_{}{}(s))$.

b) Montrons que $(AS^{-1}e = 0)$. Commençons par exprimer $(AS^{-1}e)$ :

$
AS^{-1}e = A \begin{bmatrix}
\frac{1}{s_1} & 0 & \ldots & 0 \\
0 & \frac{1}{s_2} & \ldots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \ldots & \frac{1}{s_n}
\end{bmatrix} \begin{bmatrix}
1 \\
1 \\
\vdots \\
1
\end{bmatrix}
$

Simplifions cette expression :

$AS^{-1}e = \begin{bmatrix}
A_{1,1}/s_1 + A_{1,2}/s_2 + \ldots + A_{1,n}/s_n \\
A_{2,1}/s_1 + A_{2,2}/s_2 + \ldots + A_{2,n}/s_n \\
\vdots \\
A_{m,1}/s_1 + A_{m,2}/s_2 + \ldots + A_{m,n}/s_n
\end{bmatrix}$

Chaque composante de ce vecteur est une combinaison linéaire des $s_j^{-1}$ avec des coefficients provenant de la matrice \(A\). Si ce vecteur est nul, cela signifie que chaque terme de cette combinaison linéaire est nul, ce qui implique que $AS^{-1}e = 0$.

b) Nous devons montrer que $AS^{-1}e = 0$, où \(S\) est la matrice diagonale définie par $S = \text{diag}(s_1, s_2, \ldots, s_n)$ et $e = [1, 1, \ldots, 1]^T$.

La condition centrale du centre analytique est donnée par $A^T y + s = c$. Introduisons une variable $\lambda$ associée à cette condition :

$
L(y, s, \lambda) = \sum_{j=1}^{n} \log(s_j) + \lambda^T(A^T y + s - c)
$

La condition d'optimalité nous donne :

$
\frac{\partial L}{\partial s_j} = \frac{1}{s_j} + \lambda_j = 0 \quad \text{pour } j = 1, 2, \ldots, n
$

Cela implique que $\lambda_j = -\frac{1}{s_j}$. En utilisant cette relation, la condition du centre analytique $A^T y + s - c = 0$ peut être réécrite comme :

$
A^T y - c - \frac{1}{s} = 0
$

où $\frac{1}{s}$ est une opération élément par élément. Cette équation peut être réorganisée en forme matricielle :

$
A^T y - c - S^{-1}e = 0
$

où $S^{-1}$ est la matrice inverse de \(S\). En multipliant cette équation par \(A\), nous obtenons :

$
A(A^T y - c - S^{-1}e) = 0
$

En simplifiant, cela donne $AS^{-1}e = 0$, ce qui démontre la condition souhaitée.


c) Si n = m, cela signifie que le nombre de variables est égal au nombre de contraintes. Dans ce cas, le système d'équations du centre analytique devient carré, avec autant d'équations que d'inconnues.

La condition $x \circ s = 1$ signifie que le produit élément par élément des vecteurs $x$ et $s$ est égal à 1. Cela implique que les variables d'écart $s_j$ ne peuvent pas toutes être simultanément égales à zéro, car sinon le produit $x \circ s\$ serait nul.

Ainsi, la solution du centre analytique correspond à l'intersection des plans définis par les contraintes. Chaque contrainte $A_i^T y + s_i = c_i$ définit un plan dans l'espace des variables $y$ et $s$. La condition $x \circ s = 1$ signifie que cette intersection se situe sur les bords de ces plans, où toutes les variables d'écart $s_j$ sont non nulles en même temps.

En résumé, si n = m, le centre analytique correspond à l'intersection des plans définis par les contraintes, et toutes les variables d'écart $s_j$ sont non nulles en même temps à cette solution.


## Exercice 3 

```julia
using LinearAlgebra

function center_analytic(A, b, c; α=0.01, β=0.5, kmax=50, ϵ=1e-6)
    m, n = size(A)
    
    
    x = zeros(n)
    s = ones(m)
    y = ones(m)
    
    
    for k in 1:kmax
        
        rp = s + A'*y - c[:, 1]  
        rd = A*x - s + b
        
        H = diagm(0 => s.^(-2))
        
        
        system_matrix = [zeros(m, m) A zeros(m, n); zeros(n, m) H I; A' I zeros(n, n)]
        system_rhs = [rp; rd; A*x - s + b]
        
        
        Δ = system_matrix \ system_rhs
        
        Δy = Δ[1:m]
        Δs = Δ[m+1:m+n]
        Δx = Δ[m+n+1:end]
        
       
        y += α * Δy
        s += α * Δs
        x += α * Δx
        
        
        if norm(rp) <= ϵ && norm(A*x - s + b) <= ϵ
            break
        end
    end
    
    return x
end


A = [1 1; -1 -1; 1 -1; -1 1]
b = [3, -1, 1, 1]
c = zeros(2, 1)  


result = center_analytic(A, b, c)

println("Centre analytique trouvé: ", result)
```
